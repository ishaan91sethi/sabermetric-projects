```python
# Read in dataframe
full_pitch_df = pd.read_csv("full_pitch_data.csv").drop(columns = ["Unnamed: 0"])
full_pitch_df.head()
```

# Data Cleaning/Feature Engineering

```python
# Create inning_top indicator variable
full_pitch_df["inning_top"] = (full_pitch_df["inning_topbot"] == "Top").astype("int")
# Create inning id, which is a concatenation of game_pk, inning, and inning_top indicator 
full_pitch_df["inning_id"] = [a + "_" + b + "_" + c for a, b, c in 
                               zip(full_pitch_df["game_pk"].astype("string"), full_pitch_df["inning"].astype("string"), 
                                   full_pitch_df["inning_top"].astype("string"))]


# For each inning, calculate the number of runs scored at the end of the inning by the team hitting
def end_inning_runs_scored(df):
    if (df["inning_top"].iloc[0] == 1):
        return df["post_away_score"].max()
    else:
        return df["post_home_score"].max()
runs_end_inn_df = full_pitch_df.groupby("inning_id").apply(end_inning_runs_scored)
runs_end_inn_df = runs_end_inn_df.reset_index()
runs_end_inn_df.columns = ["inning_id", "runs_end_inning"]
runs_end_inn_df
```

```python
# Turn base-runner columns from player IDs to binary variables
full_pitch_df["runner1b"] = (~full_pitch_df["on_1b"].isnull()).astype("int")
full_pitch_df["runner2b"] = (~full_pitch_df["on_2b"].isnull()).astype("int")
full_pitch_df["runner3b"] = (~full_pitch_df["on_3b"].isnull()).astype("int")

# Base-out-count state column. Format is base state (3 indicator variables for 1b, 2b, 3b), number of outs,
# and balls/strikes
full_pitch_df["state"] = [a + b + c + "_" + d + "_" + e + f for a, b, c, d, e, f in 
                          zip(full_pitch_df["runner1b"].astype("string"), full_pitch_df["runner2b"].astype("string"),
                              full_pitch_df["runner3b"].astype("string"), full_pitch_df["outs_when_up"].astype("string"),
                              full_pitch_df["balls"].astype("string"), full_pitch_df["strikes"].astype("string"))]


full_pitch_df[["runner1b", "runner2b", "runner3b", "on_1b", "on_2b", "on_3b", "outs_when_up", "balls", "strikes", "state"]].head(6)
```

```python
# Merge end of inning runs dataframe to calculate the number of runs in the rest of the half-inning for each pitch
full_pitch_df2 = pd.merge(full_pitch_df, runs_end_inn_df, how = "left", on = "inning_id")
full_pitch_df2["runs_rest_inning"] = full_pitch_df2["runs_end_inning"] - [j if i == 0 else k for i, j, k in 
                                                                          zip(full_pitch_df2["inning_top"],
                                                                              full_pitch_df2["home_score"],
                                                                              full_pitch_df2["away_score"])]

full_pitch_df2.groupby("state").aggregate({"runs_rest_inning": "mean", "game_pk": "count"}).sort_values("runs_rest_inning")
```

```python
# Create run-expectancy matrix by calculating the average number of runs scored in the rest of the inning for each inning.
# For example, using the first row in the output below, the average number of runs scored when the bases are loaded, with 
# no outs, and no balls or strikes is 2.375559. 
exp_run_matrix = full_pitch_df2.groupby("state").aggregate({"runs_rest_inning": "mean", "game_pk": "count"})
exp_run_matrix.columns = ["exp_runs", "count"]
exp_run_matrix = exp_run_matrix.reset_index()
# Add exp_runs column to dataframe
full_pitch_df2 = full_pitch_df2.merge(exp_run_matrix[["state", "exp_runs"]], how = "left", on = "state")
exp_run_matrix[exp_run_matrix["count"] > 200].sort_values("exp_runs", ascending = False).head(5)

```

Now that we have the expected number of runs for each state, we can calculate the run value of each pitch. 
What we need to do this is the expected number of runs pre-pitch (based on the state), the expected number of runs after
the pitch, and the number of runs scored during the pitch. In order to get this, we need to get the post-pitch information (i.e. the state after the pitch and the number of runs scored). I do this below.  

```python
# Goal: get out-runner-ball/strike state of next pitch
# For pitches which don't result in a ball in play it's pretty simple, can match using ab number and pitch number + 1
# For pitches which do result in a ball in play, match using ab number + 1 and pitch number = 1.
# I can do this by grouping by inning id, and then sorting first by at bat number, then by pitch number. 
# There will be missing values for the final pitch of the inning (including walk-off pitches), so I will need to 
# address this below.
# I also have to include any runs scored during the pitch. 
# End goal: have 5 columns: state pre-pitch, state after-pitch, run expectancy before pitch, run expectancy after pitch plus
# number of runs scored, and difference between prior 2 columns

# Run value of pitch = run expectancy post pitch - run expectancy pre pitch + runs scored

# Process: For pitch data from a given inning, use shift to join run expectancy after the pitch. Then find pitch run value
# by calculating: run expectancy post pitch MINUS run expectancy pre pitch PLUS runs scored.
def get_run_exp_change(df):
    #df.sort_values(["at_bat_number", "pitch_number"], inplace = True)
    # Shift dataframe back by one so we have post-pitch information. 
    post_df = df[["state", "exp_runs"]].shift(-1)
    post_df.columns = ["state_post", "exp_runs_post"]
    # Join post-pitch dataframe to original data. 
    df = pd.concat([df, post_df], axis = 1)
    #print(df.head())
    inning_top = df["inning_top"].iloc[0]
    # If top of the inning, use away scores. If bottom of the inning, use home scores.
    if (inning_top == 1):
        df["runs_scored"] = df["post_away_score"] - df["away_score"]
    else:
        df["runs_scored"] = df["post_home_score"] - df["home_score"]
    # Calculate run value of pitch. 
    df["delta_run_exp2"] = df["exp_runs_post"] - df["exp_runs"] + df["runs_scored"]
    return df

```

```python
# Sort values by inning_id, at bat number, and pitch number for above function,
full_pitch_df2.sort_values(["inning_id", "at_bat_number", "pitch_number"], inplace = True)
# Apply above function to calculate run-value per-pitch. 
full_pitch_df2 = full_pitch_df2.groupby("inning_id").apply(get_run_exp_change)
```

Now I need to address pitches that were the last of the inning. These pitches are in two categories: they end the inning,
or they end the game (walkoff). Walkoff pitches are tough because we don't know the post pitch state. For example, a walkoff single could be a ball off the wall that would normally be a double. However, because the game ends, the runner wouldn't be recorded going to second. There are two situations where walkoffs can be counted however; homeruns and walk-off walks. Walkoff pitches will have missing values except for in these two cases.

As for pitches that end the inning, this is simple. We just use the same formula from before: 

Run value of pitch = run expectancy post pitch - run expectancy pre pitch + runs scored

However, because the inning ends, the run expectancy post-pitch is simply 0, so all we need is run expectancy pre-pitch and
the number of runs scored. 

```python
# Last pitch of the inning has a missing value. Two cases: ends inning or walk off
# Case 1: Ends inning. Use the same formula as above; expected_runs_after (which is 0) - expected_runs_before + runs_scored
full_pitch_df2["delta_run_exp2"].fillna(-full_pitch_df2["exp_runs"] + full_pitch_df2["runs_scored"], inplace = True)
# Case 2: Walkoff (non-hrs and non-bbs). Make values NaN, can't tell accurately what expected_runs_after would be 
walk_off_pitches = (full_pitch_df2["inning_top"] == 0) & (full_pitch_df2["inning"] >= 9) & (full_pitch_df2["post_home_score"] > full_pitch_df2["post_away_score"])
full_pitch_df2.loc[walk_off_pitches, "delta_run_exp2"] = np.nan
# For case 3 (walkoff homeruns and BBs), we need to create a new description variable.
```

```python
# Create new description variable that has more useful descriptors. 
full_pitch_df2["description2"] = full_pitch_df2["description"]
# Replace "hit_into_play" with single, double, triple, home_run, or out.
full_pitch_df2["description2"].replace({"hit_into_play": np.nan}, inplace = True)
# Do not use events_in_play column for anything else
full_pitch_df2["events_in_play"] = [event if event in ["single", "double", "triple", "home_run"] else "out" for event in full_pitch_df2["events"]]
full_pitch_df2["description2"].fillna(full_pitch_df2["events_in_play"], inplace = True)
```

```python
##### New stuff
# Goal: For walk off homeruns and walks, I want to fill in the missing run expectancy. Using the formula:
# Run value of pitch = run expectancy post pitch - run expectancy pre pitch + runs scored
# we only need run expectancy post-pitch.
# For homeruns, this is the run expectancy of the bases empty, same number of outs of pre-pitch, and 0-0 count.

# Create dictionary mapping each state to a run expectancy
exp_runs_dict = dict(zip(exp_run_matrix["state"], exp_run_matrix["exp_runs"]))
walk_off_homeruns = (walk_off_pitches & (full_pitch_df2["description2"] == "home_run"))

# The post-homerun state in code is: "000_" + full_pitch_df2["outs_when_up"].astype("string") + "_00"
# (i.e. bases empty, same number of outs, no balls or strikes)

# Get delta_run_exp2 using our formula for run value. We get run expectancy post-pitch using the dictionary defined earlier.
full_pitch_df2.loc[walk_off_homeruns, "delta_run_exp2"] = ([exp_runs_dict[i] for i in "000_" + full_pitch_df2["outs_when_up"].astype("string") + "_00"] - full_pitch_df2["exp_runs"] 
 + full_pitch_df2["runs_scored"])


# For a walk, this is the run expectancy of the bases loaded, same number of outs pre-pitch, and 0-0 count. 
walk_off_walks = (walk_off_pitches & (full_pitch_df2["description2"] == "ball") & (full_pitch_df2["balls"] == 3))

# The post-walk state in code is: "111_" + full_pitch_df2["outs_when_up"].astype("string") + "_00"
# (i.e. bases loaded, same number of outs, empty count)
full_pitch_df2.loc[walk_off_walks, "delta_run_exp2"] = ([exp_runs_dict[i] for i in "111_" + full_pitch_df2["outs_when_up"].astype("string") + "_00"] - full_pitch_df2["exp_runs"] 
 + full_pitch_df2["runs_scored"])
```

```python
# Classify batted ball type by launch angle
full_pitch_df2.loc[full_pitch_df2["launch_angle"] < 10, "batted_ball_type"] = "gb"
full_pitch_df2.loc[full_pitch_df2["launch_angle"] >= 10, "batted_ball_type"] = "ld"
full_pitch_df2.loc[full_pitch_df2["launch_angle"] >= 25, "batted_ball_type"] = "fb"
full_pitch_df2.loc[full_pitch_df2["launch_angle"] > 50, "batted_ball_type"] = "pu"
```

Below, I create a variable calculating the average velocity of the fastest pitch in a pitcher's prior outing. Previous research
has found that the bigger the differential between a pitcher's fastball and a non-fastball (think slider, curveball, changeup),
the more effective the non-fastball is. I considered doing another rolling average within a game for each pitcher, but it 
was too computationally expensive for now. Code to calculate that is at the bottom of this notebook.

```python
# Calculate average velocity of fastest pitch in a pitcher's prior outings
def prev_fast_pitch_type(df):
    prev_pitch_df = full_pitch_df2[(full_pitch_df2["game_date"] < df["game_date"].iloc[0]) & (full_pitch_df2["pitcher"] == df["pitcher"].iloc[0])]

    df["fast_pitch_avg_velo"] = prev_pitch_df.groupby("pitch_type")[["release_speed"]].mean()["release_speed"].max()
    return df


import time
pre_time = time.time()
full_pitch_df3 = full_pitch_df2.groupby(["pitcher", "game_pk"]).apply(prev_fast_pitch_type)
print(time.time() - pre_time)
```

Below, I calculate "times through the order", or TTO. Essentially, it is tracking the amount of times a pitcher has faced
a batter within the same game. It is well known that hitters tend to benefit from facing the same pitcher multiple times in 
the game, so it is important to have this variable as a control. 

```python

# Add TTO column

def calc_tto(df):
    #df = df.sort_values(["at_bat_number", "pitch_number"])
    tto_df = df["at_bat_number"].value_counts().reset_index().sort_values("index")
    tto_df["at_bat_n"] = range(0, len(tto_df))
    tto_df["at_bat_n"] = tto_df["at_bat_n"] // 9 + 1
    df["tto"] = (tto_df["at_bat_n"].repeat(tto_df["at_bat_number"])).values
    return df


full_pitch_df3 = full_pitch_df3.groupby(["game_pk", "pitcher"], as_index = False).apply(calc_tto)
```

```python
# Zero out run values for 2 strike foul balls. A 2-strike foul ball doesn't change the state so therefore has 
# zero run value (although I've done my own research and found that hitters tend to do better the more foul balls they hit)
full_pitch_df3.loc[(full_pitch_df3["description2"] == "foul") & (full_pitch_df3["strikes"] == 2), "rv"] = 0
```

```python
# Remove rows missing any of these characteristics. Removes about 3000 rows. 
pitch_char = ["release_speed", "release_pos_x", "release_pos_z", "release_extension", "pfx_x", "pfx_z", 
              "plate_x", "plate_z", "release_spin_rate"]

missing_velo_df = full_pitch_df3[full_pitch_df3["release_speed"].isna()]

print(len(missing_velo_df))
for i in pitch_char:
    print(i)
    print(missing_velo_df[i].isna().sum() / len(missing_velo_df))

print(len(full_pitch_df3) - len(full_pitch_df3.dropna(subset = pitch_char)))
# Remove pitches with missing pitch characteristics. 
full_pitch_df3 = full_pitch_df3.dropna(subset = pitch_char)
```

```python
# Convert vertical and horizontal movement to inches for easier interpretability. Generally teams use inches instead of feet.
full_pitch_df3["pfx_z"] = 12 * full_pitch_df3["pfx_z"]
full_pitch_df3["pfx_x"] = 12 * full_pitch_df3["pfx_x"]
```
